```python
#Import Neccesary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the csv file into DF
i_df = pd.read_csv("Iris.csv")
```

```python
i_df.head()
```

```python
#check for missing values
i_df.isnull().sum()
```

```python
# Display percentage of missing values in each column
missing_percentages = (i_df.isnull().mean()*100)
print(missing_percentages)
```

```python
# Check for duplicate rows
i_df.duplicated()
```

```python
# Display data types of each column
i_df.dtypes
```

```python
# Check for outliers (basic descriptive statistics)
print("\nDescriptive statistics:")
print(i_df.describe())
```

```python
# Create histograms using seaborn
plt.figure(figsize=(12,8))
```

```python
# Histogram for Sepal Length
plt.subplot(2, 2, 1)
sns.histplot(i_df['SepalLengthCm'], bins=20, kde=True, color='blue')
plt.title('Sepal Length Distribution')
```

```python
# Histogram for Sepal Width
plt.subplot(2, 2, 2)
sns.histplot(i_df['SepalWidthCm'], bins=20, kde=True, color='green')
plt.title('Sepal Width Distribution')
```

```python
# Histogram for Petal Length
plt.subplot(2, 2, 3)
sns.histplot(i_df['PetalLengthCm'], bins=20, kde=True, color='red')
plt.title('Petal Length Distribution')
```

```python
# Histogram for Petal Width
plt.subplot(2, 2, 4)
sns.histplot(i_df['PetalWidthCm'], bins=20, kde=True, color='purple')
plt.title('Petal Width Distribution')
```

```python
# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate the average values of each feature for each species
average_values = i_df.groupby('Species').mean().reset_index()
```

```python
# Create a bar chart for each feature
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 2, 1)
sns.barplot(x='Species', y='SepalLengthCm', data=average_values, hue='Species', dodge=False, palette='Blues')
plt.title('Average Sepal Length by Species')
plt.legend([], [], frameon=False)  # This line removesthelegend
```

```python
plt.subplot(2, 2, 2)
sns.barplot(x='Species', y='SepalWidthCm', data=average_values, palette='Greens')
plt.title('Average Sepal Width by Species')
plt.show()
```

```python
plt.subplot(2, 2, 3)
sns.barplot(x='Species', y='PetalLengthCm', data=average_values, hue='Species', dodge=False, palette='Reds')
plt.title('Average Petal Length by Species')
plt.legend().remove()  # Removes the legendifnotneeded 
```

```python
plt.subplot(2, 2, 4)
sns.barplot(x='Species', y='PetalWidthCm', data=average_values, palette='Purples', dodge=False)
plt.title('Average Petal Width by Species')
plt.legend().remove()  # Remove the legend if not needed
plt.show()
```

```python
# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
